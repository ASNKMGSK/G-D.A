# 시각화 탐색 실습

```python
# 1. 필료 라이브러리 Import
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# 폰트 지정
import os
if os.name == 'nt' : 
    font_family = "Malgun Gothic"
else:
    font_family = "AppleGothic"
        
sns.set(font=font_family, rc ={"axes.unicode_minus" : False})
```

```python
# 2. 데이터 셋 생성

# tips data 로드
data = sns.load_dataset('tips')
```

```python
# 3. 시각화 탐색
# !) 상자그림(Box Plot)

# 백분위수 게싼 (percentile)
print(data['tip'].quantile([0.05,0.25,0.5,0.45,0.95]))

print("\n")
percentages = [0.05,0.25,0.5,0.75,0.95]
df = pd.DataFrame(data['tip'].quantile(percentages))
df.index = [f'{p*100}%' for p in  percentages]
print(df.transpose())
```

```python
# 한 변수에 대한 이상값 시각화
ax = (data['tip']).plot.box()
ax.set_ylabel('달러($)')
```

```python
# 상자그림을 이용한 전체 변수의 이상값 확인
plt.figure(figsize=(1,5))
sns.boxplot(data=data, orient='h')
plt.show()
```

```python
# 2) 막대 그래프

# 요일별 tip 비교
tips_sum_day= data.groupby('day').tip.sum()
tips_sum_day
```

```python
# matplotlib 막대 그래프 시각화
label = ['Thur', 'Fri','Sat','Sun']
plt.bar(label,tips_sum_day, width=0.5, align='edge')
```

```python
# Seaborn 막대 그래프 시각화
# 성별, 평균 팁 비교
sns.barplot(data= data, x= "sex", y="tip")
```

```python
# 식사 인월(size)별 남성과 여성 tip 평균 비교
## 여러 열에서 집단 묶어서 세부 집단 시각화 (hue)
plt.figure(figsize=(10,5))
sns.barplot(data =data, x="sex", y="tip", hue="size")
```

```python
# 3) 도수분포표 & 히스토그램

# 도수분포표
# 총 계산 금액 구간별 빈도수
bin_total=pd.cut(data['total_bill'], 8)
bin_total.value_counts()
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,5))

# 히스토그램
as1 = plt.subplot(1,2,1)
ax1=(data['total_bill']).plot.hist(bins=8)
ax1.set_xlabel('total_bill($)')

# 밀도그림
plt.subplot(1,2,2)
ax2 =  data['total_bill'].plot.hist(density=True, xlim=[0,60], bins=8)
data['total_bill'].plot.density(ax=ax2)
ax2.set_xlabel('total_bill($)')

plt.tight_layout()
plt.show()
```
